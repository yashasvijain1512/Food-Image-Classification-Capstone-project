# Example: Quick training + evaluation (Food-101 streaming)

```python
# This notebook demonstrates how to create a pipeline and train a baseline model
from preprocessDataset import create_data_pipeline
from train_models import build_baseline_cnn

# Create a small pipeline (use a subset for quick runs)
ds, info = create_data_pipeline(split='train', batch_size=32, img_size=224, augment=True, cache=False)
val_ds, _ = create_data_pipeline(split='validation', batch_size=32, img_size=224, augment=False)

model = build_baseline_cnn(info['num_classes'])
model.fit(ds.take(1000), validation_data=val_ds.take(100), epochs=10)

# Save and evaluate
model.save('models/baseline_quick.h5')
```
